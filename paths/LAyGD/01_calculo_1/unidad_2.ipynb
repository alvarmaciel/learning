# Problema 4
Calcula el límite indicado para cada función justificando la respuesta
mediante procedimientos algebraicos. Asimismo, acompaña el desarrollo con
una gráfica de la función y una tabla de valores donde se muestre la existencia,
o no, del límite.

## Funciones practicas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# calculo de valores de una funcion
def f(x, funcion):
    try:
        return eval(funcion)
    except ZeroDivisionError:
        return ("?")

# Crear valores para x en un rango evitando el valor indeterminad
def valores_x_factory(limite_izq, limite_der, rango, indet=None):
    valores_x = np.linspace(limite_der, limite_izq, rango)
    if indet:
        valores_x = valores_x[valores_x != indet]
    return valores_x

def table(limite_izq, limite_der, rango, function):
    x_table= valores_x_factory(limite_izq, limite_der, rango)
    y_table = f(x_table, function)
    return x_table, y_table

```

## problema 2
$\lim_{x \to 3}\frac{x-3}{x^2-9}$

```python
funcion = "(x-3)/(x**2-9)"
```

```python
limite = f(3, "1/(x+3)")
limite
```

```python
valores_x=valores_x_factory(1,5, 400, 3)
valores_y = np.array([f(x, funcion) for x in valores_x])
# Table
x_table, y_table = table(2.5, 3.5, 5, funcion)
df = pd.DataFrame({'x': x_table, 'y':y_table})

# Preparamos la gráfica
plt.figure(figsize=(8,6))
plt.plot(valores_x, valores_y, label="f(x)=(x-3)/(x^2-9)")
plt.xlim(1, 5)
plt.ylim(0, 1)

# Linea de límite
plt.axvline(x=3, color='red', linestyle='--', label="x = 3 (linea de Limite)")

# Agregamos etiquetas
plt.xlabel('x')
plt.ylabel("f(x)")
plt.title('Gráfico de la funcion \n "f(x)=(x-3)/(x^2-9)"')
plt.legend()

plt.grid(True)
plt.savefig('calc_1_uni_2_prob_4_2.png')
plt.show()
plt.close()
print(f"Tabla \n {df}")
```
